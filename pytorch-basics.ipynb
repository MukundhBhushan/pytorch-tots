imports

```python
import torch
from torch import nn
import torch.nn.functional as F
from torch.autograd import Variable
```

```python
input_size=2
print(torch.cuda.is_available())
```

training and output data

```python
training_data = Variable(torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]]))
target_data = Variable(torch.FloatTensor([[0],[1],[1],[0]]))
print(training_data,target_data)
```

Neural Network

```python
model = nn.Sequential(
    nn.Linear(input_size,100,True),
    nn.Sigmoid(),
    #nn.Dropout(0.3),
    nn.Linear(100, 1),
    nn.Sigmoid()
)
```

```python
loss_fn = nn.MSELoss()
learning_rate = 1e-4
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for i in range(1500):
    optimizer.zero_grad()  
    pred = model(training_data)
    loss = loss_fn(pred,target_data)
    print(i, loss.item())
    loss.backward()
    optimizer.step()
    
```
